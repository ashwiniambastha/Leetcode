## Add Binary

Given two binary strings a and b, return their sum as a binary string.

```python

class Solution(object):
    def addBinary(self, a, b):
        """
        :type a: str
        :type b: str
        :rtype: str
        """
        # Convert binary strings to integers (base 2)
        binary_a = int(a, 2)
        binary_b = int(b, 2)

        add = binary_a + binary_b

        if add == 0:
            return "0"

        binary = []
        quo = add
        while quo != 0:
            rem = quo % 2
            quo = quo // 2
            binary.append(str(rem))

        binary.reverse()
        return "".join(binary)
```
