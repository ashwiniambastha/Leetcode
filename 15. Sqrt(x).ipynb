```python
class Solution(object):
    def mySqrt(self, x):
        """
        :type x: int
        :rtype: int
        """
        if x == 0:
            return 0

        low = 1
        high = x
        result = 0

        while low <= high:
            mid = (low + high) // 2

            if mid * mid == x:
                return mid
            elif mid * mid < x:
                result = mid       # mid is a valid candidate, keep looking for a bigger one
                low = mid + 1
            else:
                high = mid - 1

        return result
```
