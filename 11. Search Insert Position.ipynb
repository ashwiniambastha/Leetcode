
#### Search Insert Position
Sorteted array of distinct integers and a target value, return the index if the target is found. If not, return the index where it would be if it were inserted in order.

You must write an algorithm with O(log n) runtime complexity.

```python
class Solution(object):
    def searchInsert(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: int
        """
        l=len(nums)

        for i in range(l):
            if target == nums[i]:
                return i

            if target<nums[i]:
                return i
            
            if target>nums[l-1]:
                return l

            
                
```
