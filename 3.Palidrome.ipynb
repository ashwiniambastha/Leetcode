
 Palindrome Number

Given an integer x, return true if x is a palindrome, and false otherwise.

```python
class Solution(object):
    def isPalindrome(self, x):
        """
        :type x: int
        :rtype: bool
        """
        
        
        if x<0:
            return False
        
        if x%10==0 and x!=0:
            return False
        
        reverted=0

        while x>reverted:
            digit=x%10
            reverted=reverted*10+digit
            x=x//10

        return x == reverted or x == reverted // 10
```
